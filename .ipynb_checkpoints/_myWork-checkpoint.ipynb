```python
import requests
import csv
```

```python
requests.get("https://dummyjson.com/products/")
```

```python
data = requests.get("https://dummyjson.com/products/")
```

```python
data=data.json()
```

```python
data.keys()
```

```python
data["products"]
```

```python
for i in data["products"]:
    print("ID :{}, Tiltle :{},Price :{} , brand : {}, category : {}, Stock : {}".format(i["id"],i["title"],i["price"],i["brand"],i["category"],i["stock"]))

    
```

```python
with open('data.csv', mode='w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['ID', 'Title', 'Price','brand','category','stock'])
    for i in data["products"]:
        writer.writerow([i["id"],i["title"],i["price"],i["brand"],i["category"],i["stock"]])
```

```python
ls=[]
```

```python
for i in data["products"]:
    dct={}
    dct['id']=i['id']
    dct['title']=i['title']
    dct['price']=i['price']
    dct['brand']=i['brand']
    dct['category']=i['category']
    dct['stock']=i['stock']
    ls.append(dct)
print(ls)
```

```python
field_name1=['id', 'title', 'price','brand','category','stock']
with open('output.csv',mode='w',newline='') as pushdata:
    write=csv.DictWriter(pushdata,fieldnames=field_name1)
    write.writeheader()
    write.writerows(ls)
```
